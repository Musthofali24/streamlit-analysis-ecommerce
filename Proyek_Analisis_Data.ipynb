## Data Wrangling

### Gathering Data

Pada tahap ini, kita akan mengumpulkan dan mengimpor dataset yang akan digunakan untuk analisis. Dataset yang digunakan mencakup berbagai informasi seperti data pelanggan, geolokasi, detail item pesanan, pembayaran, ulasan, pesanan, produk, serta informasi penjual. Setiap dataset ini akan dibaca dan disimpan dalam DataFrame untuk memudahkan proses analisis dan manipulasi data.

```python
# Membaca setiap dataset dan menyimpannya ke dalam DataFrame
customers_df = pd.read_csv('data/customers_dataset.csv')
geolocation_df = pd.read_csv('data/geolocation_dataset.csv')
order_items_df = pd.read_csv('data/order_items_dataset.csv')
order_payments_df = pd.read_csv('data/order_payments_dataset.csv')
order_reviews_df = pd.read_csv('data/order_reviews_dataset.csv')
orders_df = pd.read_csv('data/orders_dataset.csv')
product_category_translation_df = pd.read_csv('data/product_category_name_translation.csv')
products_df = pd.read_csv('data/products_dataset.csv')
sellers_df = pd.read_csv('data/sellers_dataset.csv')

order_items_products_df = pd.merge(order_items_df, products_df, on='product_id', how='inner')
order_items_products_df = pd.merge(order_items_products_df, product_category_translation_df, on='product_category_name', how='left')

# Menampilkan beberapa baris pertama dari setiap DataFrame untuk memastikan data berhasil dimuat
print("Customers DataFrame:")
display(customers_df)

print("Geolocation DataFrame:")
display(geolocation_df)

print("Order Items DataFrame:")
display(order_items_df)

print("Order Payments DataFrame:")
display(order_payments_df)

print("Order Reviews DataFrame:")
display(order_reviews_df)

print("Orders DataFrame:")
display(orders_df)

print("Product Category Translation DataFrame:")
display(product_category_translation_df)

print("Products DataFrame:")
display(products_df)

print("Sellers DataFrame:")
display(sellers_df)
```

### Assessing Data

Pada tahap ini, kita akan mengevaluasi kualitas data yang telah dikumpulkan untuk memastikan bahwa data siap digunakan dalam analisis. Langkah-langkah yang dilakukan pada tahap ini meliputi:

1. **Pemeriksaan Missing Values**:
    - Mengidentifikasi apakah ada nilai yang hilang (missing values) dalam setiap DataFrame.
    - Jika terdapat banyak missing values, maka perlu ditentukan strategi untuk menanganinya, seperti menghapus baris/kolom atau mengisi nilai yang hilang.

2. **Pemeriksaan Duplikasi**:
    - Memeriksa apakah ada baris duplikat dalam setiap DataFrame.
    - Duplikasi data dapat menyebabkan bias dalam hasil analisis, sehingga baris duplikat perlu dihapus.


```python
# Menyimpan hasil missing values ke dalam DataFrame
missing_values_df = pd.DataFrame({
    "DataFrame": ["Customers", "Geolocation", "Order Items", "Order Payments",
                  "Order Reviews", "Orders", "Product Category Translation", "Products", "Sellers"],
    "Missing Values": [
        customers_df.isnull().sum().sum(),
        geolocation_df.isnull().sum().sum(),
        order_items_df.isnull().sum().sum(),
        order_payments_df.isnull().sum().sum(),
        order_reviews_df.isnull().sum().sum(),
        orders_df.isnull().sum().sum(),
        product_category_translation_df.isnull().sum().sum(),
        products_df.isnull().sum().sum(),
        sellers_df.isnull().sum().sum()
    ]
})

# Menyimpan hasil duplikasi ke dalam DataFrame
duplicates_df = pd.DataFrame({
    "DataFrame": ["Customers", "Geolocation", "Order Items", "Order Payments",
                  "Order Reviews", "Orders", "Product Category Translation", "Products", "Sellers"],
    "Duplicate Rows": [
        customers_df.duplicated().sum(),
        geolocation_df.duplicated().sum(),
        order_items_df.duplicated().sum(),
        order_payments_df.duplicated().sum(),
        order_reviews_df.duplicated().sum(),
        orders_df.duplicated().sum(),
        product_category_translation_df.duplicated().sum(),
        products_df.duplicated().sum(),
        sellers_df.duplicated().sum()
    ]
})

# Menampilkan hasil dalam bentuk tabel
print("\nChecking for Missing Values and Duplicates:\n")
print("Missing Values in each DataFrame:")
display(missing_values_df)

print("\nChecking for Duplicate Rows in each DataFrame:\n")
print("Duplicate Rows in each DataFrame:")
display(duplicates_df)
```

**Insight:**
- Banyak missing values di kolom penting seperti review_comment_message.
- Ditemukan duplikasi di data geolocation, yang bisa menyebabkan bias pada analisis lokasi.

### Cleaning Data

Pada tahap ini, kita akan membersihkan data yang telah dikumpulkan untuk memastikan data tersebut siap digunakan dalam analisis. Proses pembersihan ini meliputi penanganan missing values, penghapusan duplikasi, dan konversi tipe data yang tidak sesuai.

### Langkah-langkah Pembersihan Data:

1. **Mengatasi Missing Values**:
    - Missing values dapat mempengaruhi hasil analisis jika dibiarkan. Dalam data `order_reviews`, kita menghapus baris yang tidak memiliki `order_id` atau `review_score`.
    
2. **Menghapus Duplikasi**:
    - Data yang memiliki duplikasi dapat menyebabkan bias. Pada data `geolocation`, kita menghapus semua baris duplikat untuk memastikan setiap entri lokasi adalah unik.

3. **Konversi Tipe Data**:
    - Beberapa kolom yang berisi informasi tanggal pada `orders_df` dikonversi menjadi tipe `datetime` untuk memudahkan analisis waktu, seperti menghitung durasi antara pesanan dibuat hingga diterima.

```python
# Mengatasi missing values
order_reviews_cleaned = order_reviews_df.dropna(subset=['order_id', 'review_comment_title'])

# Menghapus duplikasi di data geolocation
geolocation_cleaned = geolocation_df.drop_duplicates()

# Mengonversi tipe data tanggal di orders_df
orders_df['order_purchase_timestamp'] = pd.to_datetime(orders_df['order_purchase_timestamp'], errors='coerce')
orders_df['order_approved_at'] = pd.to_datetime(orders_df['order_approved_at'], errors='coerce')
orders_df['order_delivered_carrier_date'] = pd.to_datetime(orders_df['order_delivered_carrier_date'], errors='coerce')
orders_df['order_delivered_customer_date'] = pd.to_datetime(orders_df['order_delivered_customer_date'], errors='coerce')
orders_df['order_estimated_delivery_date'] = pd.to_datetime(orders_df['order_estimated_delivery_date'], errors='coerce')

# Menampilkan hasil pembersihan data dalam bentuk tabel
print("\nData Cleaning Completed:")

# Menampilkan 5 baris pertama dari DataFrame yang telah dibersihkan
print("\nOrder Reviews after Cleaning:")
display(order_reviews_cleaned.head())

print("\nGeolocation after Cleaning:")
display(geolocation_cleaned)

print("\nOrders Data with Correct Data Types:")
display(orders_df.head())
```

**Insight:**
- Menghapus missing values meningkatkan kualitas data, tetapi bisa mengurangi jumlah sampel.
- Mengubah format tanggal memudahkan analisis waktu seperti tren penjualan atau pengiriman.

## Visualization & Explanatory Analysis

### Pertanyaan 1:
# Visualisasi: Produk dengan Rentang Harga Paling Banyak Dibeli

Visualisasi ini bertujuan untuk melihat rentang harga produk yang paling sering dibeli oleh pelanggan. Hal ini penting untuk memahami preferensi harga dari pelanggan dan membantu dalam penentuan strategi harga dan promosi.

### Langkah-Langkah:

1. **Menggabungkan Dataset `order_items` dan `products`:**
   - Dataset `order_items` digabungkan dengan `products` untuk mendapatkan informasi harga dan kategori produk yang lebih lengkap.
   - Selain itu, dataset `product_category_translation` juga ditambahkan untuk mendapatkan nama kategori produk dalam bahasa Inggris.

2. **Membuat Rentang Harga Produk:**
   - Menggunakan `pd.cut()` untuk membagi harga produk ke dalam rentang-rentang yang sudah ditentukan, seperti 0-50 BRL, 50-100 BRL, hingga 1000-5000 BRL.

3. **Menghitung Jumlah Pembelian untuk Setiap Rentang Harga:**
   - Menghitung jumlah produk yang terjual untuk setiap rentang harga dengan menggunakan `value_counts()` dan mengurutkan berdasarkan rentang harga.

4. **Visualisasi Distribusi Rentang Harga Produk:**
   - Menggunakan `sns.barplot()` untuk menampilkan jumlah produk yang terjual pada setiap rentang harga.

### Hasil:
- Rentang harga yang paling banyak dibeli adalah `0-50 BRL`.
- Mayoritas produk yang terjual berada pada rentang harga terjangkau, yaitu di bawah 100 BRL.

```python
# Menggabungkan dataset order_items dan products untuk mendapatkan informasi harga dan kategori produk
order_items_products_df = pd.merge(order_items_df, products_df, on='product_id', how='inner')
order_items_products_df = pd.merge(order_items_products_df, product_category_translation_df, on='product_category_name', how='left')

# Menampilkan beberapa baris pertama untuk memastikan penggabungan berhasil
display(order_items_products_df.head())
print("\n")

# Membuat rentang harga produk
order_items_products_df['price_range'] = pd.cut(order_items_products_df['price'],
                                                bins=[0, 50, 100, 200, 300, 400, 500, 1000, 5000],
                                                labels=['0-50', '50-100', '100-200', '200-300', '300-400', '400-500', '500-1000', '1000-5000'])

# Menghitung jumlah produk yang terjual untuk setiap rentang harga
price_range_count = order_items_products_df['price_range'].value_counts().sort_index()

# Visualisasi distribusi rentang harga produk
plt.figure(figsize=(10, 6))
sns.barplot(x=price_range_count.index, y=price_range_count.values)
plt.title('Produk dengan Rentang Harga Paling Banyak Dibeli')
plt.xlabel('Rentang Harga (BRL)')
plt.ylabel('Jumlah Pembelian')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
print("\n")

# Penjelasan
print("Analisis Rentang Harga:")
print(f"Rentang harga paling populer adalah {price_range_count.idxmax()} BRL dengan jumlah pembelian sebanyak {price_range_count.max()}.")
print("Mayoritas produk yang terjual berada pada rentang harga terjangkau, yaitu di bawah 100 BRL.")
```

## Analisis Lanjutan (Opsional)

```python
## RFM Analysis
# Mengonversi kolom-kolom yang berisi tanggal ke format datetime
orders_df['order_purchase_timestamp'] = pd.to_datetime(orders_df['order_purchase_timestamp'], errors='coerce')
orders_df['order_approved_at'] = pd.to_datetime(orders_df['order_approved_at'], errors='coerce')
orders_df['order_delivered_carrier_date'] = pd.to_datetime(orders_df['order_delivered_carrier_date'], errors='coerce')
orders_df['order_delivered_customer_date'] = pd.to_datetime(orders_df['order_delivered_customer_date'], errors='coerce')
orders_df['order_estimated_delivery_date'] = pd.to_datetime(orders_df['order_estimated_delivery_date'], errors='coerce')

display(orders_df.dtypes)

# Menghitung jumlah nilai NaT di setiap kolom datetime
display("Jumlah nilai NaT setelah konversi:")
display(orders_df.isnull().sum())

# Menghitung Recency, Frequency, dan Monetary untuk setiap pelanggan
rfm_df = orders_df.groupby('customer_id').agg({
    'order_purchase_timestamp': lambda x: (pd.to_datetime('today') - x.max()).days,  # Recency
    'order_id': 'count',  # Frequency
    'order_estimated_delivery_date': lambda x: (x.max() - x.min()).days  # Monetary (proxy)
}).reset_index()

# Mengubah nama kolom agar lebih jelas
rfm_df.columns = ['customer_id', 'Recency', 'Frequency', 'Monetary']

# Menampilkan 5 baris pertama dari hasil RFM
print("RFM Analysis DataFrame:")
display(rfm_df.head())

## Clustering
# Membuat cluster produk berdasarkan harga
def price_cluster(price):
    if price < 50:
        return 'Low Price'
    elif 50 <= price < 200:
        return 'Medium Price'
    else:
        return 'High Price'

# Menambahkan kolom cluster harga ke dalam DataFrame
order_items_products_df['Price_Cluster'] = order_items_products_df['price'].apply(price_cluster)

# Menghitung jumlah pembelian untuk setiap cluster harga
price_cluster_count = order_items_products_df['Price_Cluster'].value_counts()

# Visualisasi jumlah pembelian untuk setiap cluster harga
plt.figure(figsize=(8, 6))
sns.barplot(x=price_cluster_count.index, y=price_cluster_count.values)
plt.title('Jumlah Pembelian Berdasarkan Cluster Harga')
plt.xlabel('Cluster Harga')
plt.ylabel('Jumlah Pembelian')
plt.grid(axis='y')
plt.show()
```
